````python
# File Handling 
import os

# OCR text extraction
from paddleocr import PaddleOCR
import pdfplumber
````

# Stage 1 - Using Paddle OCR for text extraction 

**Goal:** Extract text from images and PDFs; feed OCR output into the NLP pipeline (src/pipeline/stage1_ocr.py)

## What is OCR?

**Optical Character Recognition (OCR)** converts images of text (scanned invoices,
photos of receipts, PDF pages rendered as images) into machine-readable strings.

## PaddleOCR output structure

For each detected text region, PaddleOCR returns:
```python
[bounding_box, (text_string, confidence_score)]
# bounding_box: list of 4 corner coordinates [[x1,y1],[x2,y2],[x3,y3],[x4,y4]]
# confidence: float 0-1 (how sure the model is about this text)

Reading order is NOT guaranteed — you must sort by `(y, x)` coordinates
to reconstruct top-to-bottom, left-to-right order.

````python
def ocr_text_extraction(path : str) -> str:
    """
    Extract text from an image or text file using OCR.

    Args:
        path (str): The path to the file.

    Returns:
        str: The extracted text.
    """
    # Initialize OCR model
    ocr_model = PaddleOCR(
        use_textline_orientation=True, 
        lang='en',
    )

    # Get filetype
    filetype = os.path.splitext(path)[1].lower()
    # If it's a text file, read it directly.
    if filetype == ".txt":
        with open(path, 'r') as f:
            text = f.read()
            return text
    # If it's a PDF file, use pdfplumber to extract text.
    elif filetype == ".pdf":
        with pdfplumber.open(path) as pdf:
            text = ""
            for page in pdf.pages:
                text += page.extract_text()
            return text
    # If it's an image file, use OCR to extract text.
    elif filetype in [".jpg", ".jpeg", ".png"]:
        result = ocr_model.ocr(path)
        # Extract text and confidence scores from OCR results
        texts = result[0]['rec_texts']
        conf_scrs = result[0]['rec_scores']
        # Print scanned text and confidence scores
        for text, conf in zip(texts, conf_scrs):
            print(f"Scanned text: {text}, Confidence: {conf}")
        return text
    else:
        raise ValueError("Unsupported file type")
````

````python
# Example usage on image
image_path = "/Users/kevin/Programming/full-document-q-a-system/data/sample/test_invoice.png"
print(ocr_text_extraction(image_path))
````

````python
# Example usage on PDF
pdf_path = "/Users/kevin/Programming/full-document-q-a-system/data/sample/Profile.pdf"
print(ocr_text_extraction(pdf_path))
````

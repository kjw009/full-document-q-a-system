# Phase 2 — Cleaning + Keywords 
                                                              
**Goal**: Normalize messy text and extract the most important words. (src/pipeline/stage2_clean.py)

Common cleaning steps and why:

| Step | Example | Why |
|------|---------|-----|
| Lowercase | `"NASA"` → `"nasa"` | Unify vocabulary (for non-NER tasks) |
| Remove URLs | `"http://..."` → `""` | URLs are noisy, rarely meaningful |
| Strip punctuation | `"hello!"` → `"hello"` | Reduces vocabulary noise |
| Normalise whitespace | `"hello  world"` → `"hello world"` | Consistent tokenisation |

```python
def preprocess_text(text:str) -> str:
    """
    Normalise OCR output: fix common artefacts and whitespace issues.
    """
    # Collapse multiple spaces/tabs/newlines into a single space
    text = re.sub(r'\s+', ' ', text)
    # Remove non-ASCII characters (common OCR garbling of special characters)
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)
    # Fix hyphenated line breaks: "develop-\nment" → "development"
    text = re.sub(r'(\w)- (\w)', r'\1\2', text)
    return text.strip().lower()
```

```python
# Example usage
raw_text = "This is   an example of OCR output with  multiple   spaces,\nnewlines, and non-ASCII characters like © and ®."
clean_text = preprocess_text(raw_text)
print(f"Before cleaning: {raw_text}")
print(f"After cleaning: {clean_text}")
```
